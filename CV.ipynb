```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from pyheatmap.heatmap import HeatMap
import seaborn as sns
import matplotlib
from tqdm import tqdm

from matplotlib.colors import LogNorm

# 读取数据
data = pd.read_csv(r'1.csv', header=None)

test = data.iloc[:,0:6]
a = ['timestap','trackID','x','y','weight','hight']
test.columns = a
test['xmid']=test['x']+(test['weight']/2)
test['ymid']=test['y']-(test['hight']/2)
test['space']=test['weight']*test['hight']

test = test.query("timestap <= 500")
test
```

```python
import cv2

video = "demo-result.mp4"
result_video = "demo-result-test_finally_1.mp4"
#读取视频
cap = cv2.VideoCapture(video)
#获取视频帧率
fps_video = cap.get(cv2.CAP_PROP_FPS)
print(fps_video)
# 获取视屏总帧数
frames_num=cap.get(7)
print(frames_num)
#设置写入视频的编码格式
fourcc = cv2.VideoWriter_fourcc(*"mp4v")
#获取视频宽度
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
#获取视频高度
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))


videoWriter = cv2.VideoWriter(result_video, fourcc, fps_video, (frame_width, frame_height))

frame_id = 0

while(cap.isOpened()):
    ret, frame = cap.read()
    if ret == True:
        frame_id += 1
        test1 = test.query("timestap==@frame_id")
        for trackID in test1.trackID.unique():
            test2 = test1.query("trackID==@trackID")
            left_x_up = test2['x']
            left_y_up = test2['y']
            right_x_down = test2['x'] + test2['weight']
            right_y_down = test2['y'] + test2['hight']
            #文字坐标
            word_x = left_x_up + 5
            word_y = left_y_up + 25
            cv2.rectangle(frame, (left_x_up, left_y_up), (right_x_down, right_y_down), (55,255,155), 5)
            cv2.putText(frame, 'trackID_%s' %trackID, (word_x, word_y), cv2.FONT_HERSHEY_SIMPLEX, 1, (55,255,155), 2)
        videoWriter.write(frame)
    else:
        videoWriter.release()
        break


```
